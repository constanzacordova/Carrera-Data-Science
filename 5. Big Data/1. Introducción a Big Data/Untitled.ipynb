```python
#Filter

def get_jovenes(x):
    edad = x[0]
    if edad >= 18 and edad <=24:
        return True
    return False

list_jovenes = list(filter(get_jovenes, random_db))

print(list_jovenes)
```

```python
#Reduce 
import functools

#lista solo con edades
age_jovenes = list(map(lambda x: x[0], list_jovenes))
print(age_jovenes)
```

```python
sum_age = functools.reduce(lambda x,y : x+y, age_jovenes, 0)
print(sum_age)
```

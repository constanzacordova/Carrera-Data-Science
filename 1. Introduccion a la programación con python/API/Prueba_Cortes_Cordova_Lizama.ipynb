### 1. Una función request que debe:

Recibir como parámetro la URL y el api_key

Retrornar el diciconario con los resultados

Tip: Se puede seguir el paso a paso explicado en la lectura en el capítulo API con autenticación e ingresar al Portal de la Nasa para autenticarse.


```python
# Primero importamos librería
import requests
```

```python
# Crear función request 

def custom_request(url, api_key):
 
    # Considerando que el parámetro api_key es un string, se crea un diccionario para parametro que se usará con el método get
    params = {}
    params['api_key'] = api_key

    return requests.get(url, params).json()
```

```python
#Evaluar función: se evalúa la función entregando como argumento la url de la nasa y la api_key

url_nasa = 'https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/latest_photos'
key_nasa = 'hXG2muzvMKng2HElfkIUsju4bFLMShhAihCHnbLw'

custom_request(url_nasa, key_nasa)
```

### 2. Una función llamada build_web_page que debe:

Recibir como parámetro el diccionario de respuesta con todos los datos de las fotos.

Construir una página web que muestre las fotos.


```python
# Crear función build_web_page

def build_web_page(diccionario):
    html_i = '<html>\n<head>\n</head>\n<body>\n<ul>\n' #html inicio
    html_f = "</ul>\n</body>\n</html>"                 #html final
    img = ''                                           #html para la lista de imagenes

    for imagen in diccionario['latest_photos'][0:25]:  #recorremos el diccionario en un rango de 25
        img += "\t<li><img src=\"{}\"></li>\n".format(imagen["img_src"])

    html = html_i + img + html_f

    with open("imgenes_nasa.html", "w") as f: 
        f.write(html)
        
```

```python
#Evaluar función con diccionario de nasa

diccionario_nasa = custom_request(url_nasa, key_nasa)

build_web_page(diccionario_nasa)
```

### Bonus
Crear la función photos_count que reciba el diccionario de respuesta, y
devuelva un nuevo diccionario con el nombre de la cámara y la cantidad de fotos.


```python
from itertools import groupby

def photos_count(diccionario):
    
    lista_camaras = []
    n = len(diccionario['latest_photos'])  #cantidad de imagenes que contiene el diccionario

    for i in range(n): #recorremos el diccionario en el rando de n elementos
        lista_camaras.append(diccionario['latest_photos'][i]['camera']['full_name']) #Agregamos a lista_camaras el nombre de la camara 

    lista_camaras.sort()  #Ordenamos la lista 
 
    camaras = {k: len(list(g)) for k, g in groupby(lista_camaras)} 
    
    return camaras

photos_count(diccionario_nasa)
```
